```python
# Convertir la columna 'Date/Time' a formato datetime
df_test['Date/Time'] = pd.to_datetime(df_test['Date/Time'])

# Ordenar por 'Visitor ID', 'Visit ID' y 'Date/Time' para asegurarse de que los pasos estén en el orden correcto
df = df_test.sort_values(by=['Visitor ID', 'Visit ID', 'Date/Time'])

# Calcular las diferencias de tiempo entre los pasos
df['Time_Diff'] = df.groupby(['Visitor ID', 'Visit ID'])['Date/Time'].diff()

# Eliminar los valores nulos que aparecen para los primeros pasos
df = df.dropna(subset=['Time_Diff'])

# Filtrar los datos por cada transición y calcular la media
time_diffs_df_t = {}
steps = ['start', 'step_1', 'step_2', 'step_3', 'confirm']

for i in range(len(steps) - 1):
    step_from = steps[i]
    step_to = steps[i + 1]
    mask = (df['Process Step'] == step_to) & (df.groupby(['Visitor ID', 'Visit ID'])['Process Step'].shift() == step_from)
    mean_diff = df[mask]['Time_Diff'].mean()
    time_diffs_df_t[f'{step_from}_to_{step_to}'] = mean_diff

# Mostrar los resultados
for transition, mean_time in time_diffs_df_t.items():
    print(f'Tiempo medio de {transition}: {mean_time}')
```

# Comparacion de porcentaje de error, total de transiciones por paso

```python
import pandas as pd
import matplotlib.pyplot as plt

# Función para calcular el total de transiciones y errores por paso
def calculate_step_errors(df):
    # Convertir la columna 'Date/Time' a formato datetime
    df['Date/Time'] = pd.to_datetime(df['Date/Time'])

    # Ordenar por 'Visitor ID', 'Visit ID' y 'Date/Time'
    df = df.sort_values(by=['Visitor ID', 'Visit ID', 'Date/Time'])

    # Crear una columna con el paso anterior
    df['Previous Step'] = df.groupby(['Visitor ID', 'Visit ID'])['Process Step'].shift()

    # Definir el orden correcto de los pasos
    correct_order = {
        'start': 'step_1',
        'step_1': 'step_2',
        'step_2': 'step_3',
        'step_3': 'confirm'
    }

    # Contar los errores
    df['Is_Error'] = df.apply(lambda row: row['Process Step'] != correct_order.get(row['Previous Step'], row['Process Step']), axis=1)

    # Inicializar contadores
    step_counts = {step: {'total': 0, 'errors': 0} for step in list(correct_order.keys()) + ['confirm']}

    df['Is_Error'] = df.apply(lambda row: row['Process Step'] != correct_order.get(row['Previous Step'], row['Process Step']), axis=1)


    
        
        # Contar transiciones hacia `confirm`
        
    # Contar total de transiciones y errores por paso
    for index, row in df.iterrows():
        prev_step = row['Previous Step']
        current_step = row['Process Step']
        
        # Contar las transiciones correctas hacia `confirm`
        if current_step == 'confirm' and prev_step == 'step_3':
            step_counts['step_3']['total'] += 1
            if row['Is_Error']:
                step_counts['step_3']['errors'] += 1
        # Contar las demás transiciones
        elif prev_step in step_counts:
            step_counts[prev_step]['total'] += 1
            if row['Is_Error']:
                step_counts[prev_step]['errors'] += 1
                
        if current_step == 'confirm':
            step_counts['confirm']['total'] += 1
            if row['Is_Error']:
                step_counts['confirm']['errors'] += 1
        # Contar las demás transiciones
        elif prev_step in step_counts:
            step_counts[prev_step]['total'] += 1
            if row['Is_Error']:
                step_counts[prev_step]['errors'] += 1

    # Calcular el porcentaje de error por paso
    for step in step_counts:
        total = step_counts[step]['total']
        errors = step_counts[step]['errors']
        percentage_error = (errors / total) * 100 if total > 0 else 0
        step_counts[step]['percentage_error'] = percentage_error

    return step_counts

df1 = df_test
df2 = df_control

# Calcular los errores y transiciones por paso para cada DataFrame
step_counts1 = calculate_step_errors(df1)
step_counts2 = calculate_step_errors(df2)

# Crear la gráfica de barras
fig, ax = plt.subplots(figsize=(12, 8))

# Datos para la gráfica
labels = list(step_counts1.keys())
total_transitions1 = [step_counts1[step]['total'] for step in labels]
total_errors1 = [step_counts1[step]['errors'] for step in labels]
percentage_errors1 = [step_counts1[step]['percentage_error'] for step in labels]

total_transitions2 = [step_counts2[step]['total'] for step in labels]
total_errors2 = [step_counts2[step]['errors'] for step in labels]
percentage_errors2 = [step_counts2[step]['percentage_error'] for step in labels]

x = range(len(labels))

# Crear las barras
bar_width = 0.35
ax.bar(x, total_transitions1, width=bar_width, label='Total de Transiciones Test', align='center', color='b')
ax.bar([i + bar_width for i in x], total_transitions2, width=bar_width, label='Total de Transiciones Control', align='center', color='g')

ax.bar(x, total_errors1, width=bar_width, label='Total de Errores Test', align='edge', color='r')
ax.bar([i + bar_width for i in x], total_errors2, width=bar_width, label='Total de Errores Control', align='edge', color='y')

# Añadir el porcentaje de error sobre las barras
for i in range(len(labels)):
    ax.text(i, total_transitions1[i] + 1, f'{percentage_errors1[i]:.2f}%', ha='center', color='b')
    ax.text(i + bar_width, total_transitions2[i] + 1, f'{percentage_errors2[i]:.2f}%', ha='center', color='g')

    ax.text(i, total_errors1[i] + 1, f'{percentage_errors1[i]:.2f}%', ha='center', color='r')
    ax.text(i + bar_width, total_errors2[i] + 1, f'{percentage_errors2[i]:.2f}%', ha='center', color='y')

# Añadir etiquetas y título
ax.set_xlabel('Pasos del Proceso')
ax.set_ylabel('Cantidad')
ax.set_title('Comparación de Errores y Transiciones entre DataFrames por Paso')
ax.set_xticks([i + bar_width / 2 for i in x])
ax.set_xticklabels(labels)
ax.legend()

# Mostrar la gráfica
plt.show()
```

## Calculo de error de cada df

```python
# Convertir la columna 'Date/Time' a formato datetime
df_test['Date/Time'] = pd.to_datetime(df_test['Date/Time'])

# Ordenar por 'Visitor ID', 'Visit ID' y 'Date/Time' para asegurarse de que los pasos estén en el orden correcto
df = df_test.sort_values(by=['Visitor ID', 'Visit ID', 'Date/Time'])

# Crear una columna con el paso anterior
df['Previous Step'] = df.groupby(['Visitor ID', 'Visit ID'])['Process Step'].shift()

# Definir el orden correcto de los pasos
correct_order = {
    'start': 'step_1',
    'step_1': 'step_2',
    'step_2': 'step_3',
    'step_3': 'confirm'
}

# Contar los errores
df['Is_Error'] = df.apply(lambda row: row['Process Step'] != correct_order.get(row['Previous Step'], row['Process Step']), axis=1)

# Calcular el número total de transiciones y el número de errores
total_transitions_t = len(df) - df['Previous Step'].isna().sum()  # Excluir las primeras filas que no tienen paso anterior
total_errors_t = df['Is_Error'].sum()

# Calcular el porcentaje de error
percentage_error_t = (total_errors_t / total_transitions_t) * 100

print(f'Total de transiciones: {total_transitions_t}')
print(f'Total de errores: {total_errors_t}')
print(f'Porcentaje de error: {percentage_error_t:.2f}%')
```
